# P2.369게임을 함수로 만들어 보기

```python

# 함수 정의
def intro():
    print('''
    재미나는 3!6!9! 게임!

    1~난수까지 숫자를 차례대로 출력하되 숫자에 3,6,9가 있으면 그 개수만큼 '짝' 을 출력하고, 그렇지 않다면 숫자를 출력한다.
    ''')
    return

def genRand():
    import random
    rnd=random.randint(1,100)
    return rnd

def count_369(n):
    cnt=0
    for num_txt in str(n):
        if num_txt in '369':
            cnt=cnt+1
    return cnt
def printClap(n,cnt):
    if cnt>0:
        print(n,'---->','짝'*cnt)
    else:
        print(n,'---->',n)
    return

#프로그램 실행
intro()
RandN = genRand()
print('발생된 난수 : ',RandN)

for num in range(1,RandN+1):
    count=count_369(num)
    printClap(num,count)
        
```

# P3.재귀함수

### P3-1.팩토리얼을 재귀함수로 구현

```python
#함수
def fact(n):
    if n==1:
        return 1
    num=n* fact(n-1)
    return num

#실행
for i in range(1,4):
    num=int(input('숫자 입력 : '))
    print(num,'! = ',fact(num))
```

### P3-2.피보나치 수열을 재귀함수로 구현

```python
#함수 정의
def fibo(n):
    if n<=2:
        return 1
    return fibo(n-1)+fibo(n-2)

#함수 실행
for i in range(1,3):
    n = int(input('숫자 입력 : '))
    print('피보나치수열',n,'번째 수 : ',fibo(n))
```

# P4.로또프로그램

```python
#함수정의

def SetLottoNum():
    import random
    randomSet = list(range(1,46))
    random.shuffle(randomSet)
    lottoNumList = randomSet[:6]
    return lottoNumList

def GetLottoNum():
    playerNumList = []
    while len(playerNumList)<6:
        num = int(input('숫자입력: '))
        if num not in playerNumList:
            playerNumList.append(num)
        else:
            print('중복입니다. 다시 입력하세요.')
            
    return playerNumList

def LottoResult(lottoNumList,playerNumList):
    cnt=0
    correctList=[]
    for c in playerNumList:
        if int(c) in lottoNumList:
            cnt= cnt+1
            correctList.append(c)
    return cnt,correctList

#함수실행



rewardList = [0,0,0,0,1000,5000,10000]
nowReward =0


while True:
    playerList=[]
    thisLottoList = SetLottoNum()
    
    print()
    print('로또번호가 선택됐습니다.')
    print('사용자 번호를 중복없이 하나씩 입력해주세요.')
    
    playerList = GetLottoNum()
    
    
    
    score,thisCorrectList = LottoResult(thisLottoList,playerList)
    nowReward = nowReward + rewardList[score]
    print('로또 번호는',thisLottoList,'입니다.')
    print('사용자가 선택한 번호는',playerList,'입니다.')
    print('일치한 숫자의 갯수는',score,'이고')
    print('일치한 숫자는',thisCorrectList,'입니다.')
    print('현재까지 상금은',nowReward,'입니다.')

    retry = input('계속 하시겠습니까? y/n ')
    if retry =='n':
        print('총상금은',nowReward,'입니다.')
        break
        
```
