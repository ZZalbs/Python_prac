# P1. 다음 순서에 따라 계수 정렬을 구현하시오.

```python
import random

def RandGet():
    ranList = list(range(1,101))
    random.shuffle(ranList)
    return ranList[:10]

def MakeList(ranList):
    listA = []
    for i in range(0,1000):
        listA.append(ranList[random.randint(0,9)])
    return listA

def MinMax(listA):
    minNum = 10000
    maxNum = -1
    for i in range(0,len(listA)):
        if listA[i]<minNum:
            minNum = listA[i]
        if listA[i]>maxNum:
            maxNum = listA[i]
    return minNum,maxNum


##메인
list_unordered = MakeList(RandGet())
print(list_unordered)
start,end = MinMax(list_unordered)

num=0
listC =[]
listB =[]
for i in range(0,end+1):
    listC.append(0)

while (num < len(list_unordered)):
    listC[list_unordered[num]] = (listC[list_unordered[num]]+1)
    num=num+1

num=0
while num<len(listC):
    while num<len(listC) and listC[num] >0:
        listB.append(num)
        listC[num]=listC[num]-1
    num=num+1
print(listB)
```

# P3. 영한사전 프로그램

```python
dictStr = '''better;더나은
tomorrow;내일
always;언제나
best;최고의
you;당신'''

file = open('dictionary.txt','w',encoding='utf-8')
file.write(dictStr)
file.close()



word = input()
check = open('dictionary.txt','r',encoding='utf-8')
a = check.readline()
wordin=False
while a != '':
    if(word == a.rstrip().split(';')[0]):
        
        print("'"+a.rstrip().split(';')[1]+"'이라는 뜻을 가진 단어입니다.")
        wordin=True
        break
    a = check.readline()
    
if(wordin==False):
    print('단어가 없습니다.')
file.close()
```

# P4. 성적처리

```python
file = open('grade.txt','r',encoding='utf-8')
scores = file.readlines()
file.close()

clean_score = []

for i in scores:
    clean_score.append(i.rstrip().split('\t'))

#print(clean_score)
avg_mid,avg_final,avg_work,avg_attend = 0,0,0,0

for line in clean_score[1:]:
    avg_mid=avg_mid+int(line[1])
    avg_final=avg_final+int(line[2])
    avg_work=avg_work+int(line[3])
    avg_attend=avg_attend+int(line[4])
    
num_student = len(clean_score)-1
avg_mid = round(avg_mid/num_student,4)
avg_final = round(avg_final/num_student,4)
avg_work = round(avg_work/num_student,4)
avg_attend = round(avg_attend/num_student,4)

print('학생수:',num_student)
print('중간평균:',avg_mid)
print('기말평균:',avg_final)
print('과제평균:',avg_work)
print('출석평균:',avg_attend)
```
